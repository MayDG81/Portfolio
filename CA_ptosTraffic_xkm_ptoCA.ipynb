```python
# distancia entre dos puntos a partir de coordenadas (fórmula del Semiverseno)
def haversine(lon1, lat1, lon2, lat2):

    lon1 = np.radians(lon1)
    lat1 = np.radians(lat1)
    lon2 = np.radians(lon2)
    lat2 = np.radians(lat2)

    r = 6371


    dlon = np.subtract(lon2, lon1)
    dlat = np.subtract(lat2, lat1)

    a = np.add(np.power(np.sin(np.divide(dlat, 2)), 2),
               np.multiply(np.cos(lat1),
                           np.multiply(np.cos(lat2),
                                       np.power(np.sin(np.divide(dlon, 2)), 2))
                           )
              )
    c = np.multiply(2, np.arcsin(np.sqrt(a)))

    return c*r
```

***Distancias a estaciones de CA de interés***



```python
lat_RC, long_RC = 40.45167, -3.67722
lat_B, long_B = 40.47472, -3.57500
lat_CC, long_CC = 40.44528, -3.70639
lat_R, long_R = 40.40972, -3.68556
lat_PE, long_PE = 40.38470, -3.71860
```

```python
pto_t['dist_RC'] = haversine(pto_t.longitud, pto_t.latitud, long_RC, lat_RC)
pto_t['dist_B'] = haversine(pto_t.longitud, pto_t.latitud, long_B, lat_B)
pto_t['dist_CC'] = haversine(pto_t.longitud, pto_t.latitud, long_CC, lat_CC)
pto_t['dist_R'] = haversine(pto_t.longitud, pto_t.latitud, long_R, lat_R)
pto_t['dist_PE'] = haversine(pto_t.longitud, pto_t.latitud, long_PE, lat_PE)
```

```python
pto_t
```

```python
puntos_RC = pto_t[pto_t.dist_RC <= 2].id.reset_index(drop=True)
print(f'# puntos de tráfico RC: {len(puntos_RC)}')

puntos_B = pto_t[pto_t.dist_B <= 2].id.reset_index(drop=True)
print(f'# puntos de tráfico B: {len(puntos_B)}')

puntos_CC = pto_t[pto_t.dist_CC <= 2].id.reset_index(drop=True)
print(f'# puntos de tráfico CC: {len(puntos_CC)}')

puntos_R = pto_t[pto_t.dist_R <= 2].id.reset_index(drop=True)
print(f'# puntos de tráfico R: {len(puntos_R)}')

puntos_PE = pto_t[pto_t.dist_PE <= 2].id.reset_index(drop=True)
print(f'# puntos de tráfico PE: {len(puntos_PE)}')
```

```python
ptos_x_est_CA_2km = pd.DataFrame()
ptos_x_est_CA_2km['ptos_RC'] = puntos_RC
ptos_x_est_CA_2km['ptos_B'] = puntos_B
ptos_x_est_CA_2km['ptos_CC'] = puntos_CC
ptos_x_est_CA_2km['ptos_R'] = puntos_R
ptos_x_est_CA_2km['ptos_PE'] = puntos_PE

ptos_x_est_CA_2km.to_csv('/content/drive/MyDrive/madrid_traffic_data/ptos_x_est_CA_2km.csv')

ptos_x_est_CA_2km
```
